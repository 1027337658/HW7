### Q1: Derivation of $p(\boldsymbol{\beta}|\mathbf{y},\mathbf{X},\Sigma = \sigma^{2}I)$ and posterior distribution for $\sigma^{2}$

#### 1. Derivation of $p(\boldsymbol{\beta}|\mathbf{y},\mathbf{X},\Sigma = \sigma^{2}I)$

Given the linear regression model likelihood $p(\mathbf{y}|\mathbf{X},\boldsymbol{\beta},\Sigma)$ is proportional to $\exp\left(-\frac{1}{2}(\mathbf{y}-\mathbf{X}\boldsymbol{\beta})^{\top}\Sigma^{-1}(\mathbf{y}-\mathbf{X}\boldsymbol{\beta})\right)$. Because we have $\Sigma=\sigma^{2}I$, that is $\Sigma^{-1}=\frac{1}{\sigma^{2}}I$ so that we have the likelihood is equally proportional to $\exp\left(-\frac{1}{2\sigma^{2}}(\mathbf{y}-\mathbf{X}\boldsymbol{\beta})^{\top}(\mathbf{y}-\mathbf{X}\boldsymbol{\beta})\right)$

Also, given the  prior distribution of $\boldsymbol{\beta}$ is $p(\boldsymbol{\beta})\sim MVN (\boldsymbol{\beta}_{0},\boldsymbol{\Sigma}_{\beta})$, that is $p(\boldsymbol{\beta}) \propto \exp\left(-\frac{1}{2}(\boldsymbol{\beta}-\boldsymbol{\beta}_{0})^{\top}\boldsymbol{\Sigma}_{\beta}^{-1}(\boldsymbol{\beta}-\boldsymbol{\beta}_{0})\right)$

Using Bayes' theorem, we have the posterior distribution as following:

$p(\boldsymbol{\beta}|\mathbf{y},\mathbf{X},\Sigma)\propto p(\mathbf{y}|\mathbf{X},\boldsymbol{\beta},\Sigma)p(\boldsymbol{\beta})$

Using the previous two conditions and substitute them into the above Bayes' theorem, we have:

\[
\begin{align*}
p(\boldsymbol{\beta}|\mathbf{y},\mathbf{X},\Sigma)&\propto\exp\left(-\frac{1}{2\sigma^{2}}(\mathbf{y}-\mathbf{X}\boldsymbol{\beta})^{\top}(\mathbf{y}-\mathbf{X}\boldsymbol{\beta})\right)\times \exp\left(-\frac{1}{2}(\boldsymbol{\beta}-\boldsymbol{\beta}_{0})^{\top}\boldsymbol{\Sigma}_{\beta}^{-1}(\boldsymbol{\beta}-\boldsymbol{\beta}_{0})\right)\\
&\propto\exp\left(-\frac{1}{2\sigma^{2}}(\mathbf{y}^{\top}\mathbf{y}- 2\boldsymbol{\beta}^{\top}\mathbf{X}^{\top}\mathbf{y}+\boldsymbol{\beta}^{\top}\mathbf{X}^{\top}\mathbf{X}\boldsymbol{\beta})\right)\times\exp\left(-\frac{1}{2}(\boldsymbol{\beta}^{\top}\boldsymbol{\Sigma}_{\beta}^{-1}\boldsymbol{\beta}-2\boldsymbol{\beta}^{\top}\boldsymbol{\Sigma}_{\beta}^{-1}\boldsymbol{\beta}_{0})\right)\\
&\propto\exp\left(-\frac{1}{2}\left(\boldsymbol{\beta}^{\top}\left(\frac{1}{\sigma^{2}}\mathbf{X}^{\top}\mathbf{X}+\boldsymbol{\Sigma}_{\beta}^{-1}\right)\boldsymbol{\beta}-2\boldsymbol{\beta}^{\top}\left(\frac{1}{\sigma^{2}}\mathbf{X}^{\top}\mathbf{y}+\boldsymbol{\Sigma}_{\beta}^{-1}\boldsymbol{\beta}_{0}\right)\right)\right)
\end{align*}
\]

Thus, the above is a  $MVN\left(\boldsymbol{\mu},\boldsymbol{\Sigma}\right)$ where we have: $\boldsymbol{\mu}=\left(\frac{1}{\sigma^{2}}\mathbf{X}^{\top}\mathbf{X}+\boldsymbol{\Sigma}_{\beta}^{-1}\right)^{-1}\left(\frac{1}{\sigma^{2}}\mathbf{X}^{\top}\mathbf{y}+\boldsymbol{\Sigma}_{\beta}^{-1}\boldsymbol{\beta}_{0}\right)$ and $\boldsymbol{\Sigma}=\left(\frac{1}{\sigma^{2}}\mathbf{X}^{\top}\mathbf{X}+\boldsymbol{\Sigma}_{\beta}^{-1}\right)^{-1}$


#### 2. Posterior distribution for $\sigma^{2}$

Firstly, we have the likelihood function for the linear regression model is as following:

$p(\mathbf{y}|\mathbf{X},\boldsymbol{\beta},\sigma^{2})\propto\left(\sigma^{2}\right)^{-\frac{n}{2}}\exp\left(-\frac{1}{2\sigma^{2}}(\mathbf{y}-\mathbf{X}\boldsymbol{\beta})^{\top}(\mathbf{y}-\mathbf{X}\boldsymbol{\beta})\right)$

Also, given the prior distribution of $\sigma^{2}$ is an Inverse-Gamma($\alpha^{*}$, $\beta^{*}$), we have:

$p(\sigma^{2})\propto\left(\sigma^{2}\right)^{-\alpha^{*}- 1}\exp\left(-\frac{\beta^{*}}{\sigma^{2}}\right)$

Using Bayes' theorem, we have the posterior distribution is as following:

$p(\sigma^{2}|\mathbf{y},\mathbf{X},\boldsymbol{\beta})\propto p(\mathbf{y}|\mathbf{X},\boldsymbol{\beta},\sigma^{2})p(\sigma^{2})$

substitute the expressions based on Bayes' theorem, we have:

\[
\begin{align*}
p(\sigma^{2}|\mathbf{y},\mathbf{X},\boldsymbol{\beta})&\propto\left(\sigma^{2}\right)^{-\frac{n}{2}}\exp\left(-\frac{1}{2\sigma^{2}}(\mathbf{y}-\mathbf{X}\boldsymbol{\beta})^{\top}(\mathbf{y}-\mathbf{X}\boldsymbol{\beta})\right)\times\left(\sigma^{2}\right)^{-\alpha^{*}-1}\exp\left(-\frac{\beta^{*}}{\sigma^{2}}\right)\\
&\propto\left(\sigma^{2}\right)^{-\left(\alpha^{*}+\frac{n}{2}+ 1\right)}\exp\left(-\frac{1}{\sigma^{2}}\left(\beta^{*}+\frac{1}{2}(\mathbf{y}-\mathbf{X}\boldsymbol{\beta})^{\top}(\mathbf{y}-\mathbf{X}\boldsymbol{\beta})\right)\right)\\
&\propto\left(\sigma^{2}\right)^{-\left(\alpha^{*}+\frac{n}{2}\right)- 1}\exp\left(-\frac{1}{\sigma^{2}}\left(\beta^{*}+\frac{1}{2}(\mathbf{y}-\mathbf{X}\boldsymbol{\beta})^{\top}(\mathbf{y}-\mathbf{X}\boldsymbol{\beta})\right)\right)
\end{align*}
\]

Obviously, the above is also an inverse - gamma distribution, thus, it is Inverse-Gamma($\alpha,\beta$) where the parameters are: $\alpha=\alpha^{*}+\frac{n}{2}$ and $\beta=\beta^{*}+\frac{1}{2}(\mathbf{y}-\mathbf{X}\boldsymbol{\beta})^{\top}(\mathbf{y}-\mathbf{X}\boldsymbol{\beta})$



### Q2: perform Bayesian Linear Regression

I divided this problem into four main steps:


1. Data Reading and Pre - processing: the source data comes from kaggle with url: https://www.kaggle.com/code/zabihullah18/car-price-prediction?select=CarPrice_Assignment.csv which is a car price data, this problem use car price as dependent variable, and related **numeric predictors** only from the raw data for an easier solution

2. Building the Bayesian Linear Regression Model

- use pm.Normal define the prior distributions of the regression coefficients betas
- use pm.HalfNormal to define the prior distribution of sigma
- use pm.Normal to define the likelihood function

3. Use pm.sample to perform Model Sampling
4. Inference and MCMC Diagnostics: use plot trace to assess convergence

```python
import pandas as pd
import numpy as np
import pymc as pm
import matplotlib.pyplot as plt
import arviz as az


# 1. Data Reading and Pre - processing
data = pd.read_csv('CarPrice_Assignment.csv')

# Extract numerical variables
numeric_columns = data.select_dtypes(include=[np.number]).columns

X = data[numeric_columns.drop('price')].values
y = data['price'].values

# 2. Build Bayesian Linear Regression Model
with pm.Model() as LR:
    betas = pm.Normal('betas', mu=0, sigma=1, shape=X.shape[1])
    beta0 = pm.Normal('beta0', mu=0, sigma=1)
    sigma = pm.HalfNormal('sigma', sigma=1)
    y = pm.Normal('y', mu= beta0 + X@betas, sigma=sigma, observed=y)
```

```python
# 3. Model Sampling
with LR:
    idata = pm.sample()
```

```python
# 4. Inference and diagnostics

# Plot the trace plot of parameters
fig, axes = plt.subplots(3, 2, figsize=(12, 8))

# Plot the traces and densities
az.plot_trace(idata, var_names=['betas', 'beta0', 'sigma'], axes=axes)

plt.tight_layout()
plt.show()
```

Observations: it can be seen from bayesian posterior distributions and the race mcmc diagnostics plots that the parameters are converged, the inferenced could be reliable.



For Q2 extra, using poisson distribution, the link is exp().

```python
from sklearn.preprocessing import StandardScaler
X = data[numeric_columns.drop('price')].values
y2 = data['price'].values
# Data standardization
scaler_X = StandardScaler()
X_scaled = scaler_X.fit_transform(X)
with pm.Model() as GLM:
    betas = pm.Normal('betas', mu=0, sigma=1, shape=X_scaled.shape[1])
    beta0 = pm.Normal('beta0', mu=0, sigma=1)
    y2 = pm.Poisson('y2', mu=pm.math.exp(beta0 + pm.math.dot(X_scaled, betas)), observed=y2)
with GLM:
    idata2 = pm.sample()
```

```python
#plots
fig, axes = plt.subplots(2, 2, figsize=(12, 8))
az.plot_trace(idata2, var_names=['betas', 'beta0'], axes=axes)
plt.tight_layout()
plt.show()
```

Observations: note that using poisson for this data, scaling of data is needed due to exp(), the transformations might be too large for computation, at last, the results show the convergence is appropriate for using poisson.

### Q3: perform robust Bayesian Linear Regression

I divided this problem into four main steps:


1. Data Reading and Pre - processing: the source data comes from kaggle with url: https://www.kaggle.com/code/zabihullah18/car-price-prediction?select=CarPrice_Assignment.csv which is a car price data, this problem use car price as dependent variable, and related **numeric predictors** only from the raw data for an easier solution, also, in this problem, some outliers are added into the data.

2. Building the robust Bayesian Linear Regression Model
3. Identify outliers based on $\tau_i$
4. Calculate Influence
5. Inference and MCMC Diagnostics: use plot trace to assess convergence

```python
# Load the dataset
data = pd.read_csv('CarPrice_Assignment.csv')
numeric_columns = data.select_dtypes(include=[np.number]).columns
X = data[numeric_columns.drop('price')].values
y = data['price'].values
# add some outliers
outlier_indices = np.random.choice(len(y), 5, replace=False)
y_outliers = y.copy()
y_outliers[outlier_indices] *= 2  # make them big enough to be outliers
with pm.Model() as robust_model:
    beta = pm.Normal('beta', mu=0, sigma=10, shape=X.shape[1])
    tau = pm.HalfNormal('tau', sigma=10)
    nu = pm.Gamma('nu', alpha=2, beta=0.1)  # Prior for nu
    # Hierarchical prior
    tau_i = pm.Gamma('tau_i', alpha=nu / 2, beta=nu / 2, shape=len(y_outliers))
    y_outliers = pm.Normal('y_outliers', mu=X@beta, tau=tau * tau_i, observed=y_outliers)
with robust_model:
    idata3 = pm.sample(4000, target_accept = 0.95, cores = 2)
```

```python
# Identify outliers
tau_i = idata3.posterior['tau_i'].values
tau_i_means = tau_i.mean(axis=(0, 1))
outlier_threshold = np.percentile(tau_i_means, 10)
outlier_i = tau_i_means < outlier_threshold
def hat_matrix_diag(X, D):
    XtDX = X.T @ D @ X
    try:
        inv_XtDX = np.linalg.inv(XtDX)
    except np.linalg.LinAlgError:
        inv_XtDX = np.linalg.pinv(XtDX)
    H = X @ inv_XtDX @ X.T
    return np.diag(H)
# posterior distributions of sigma_i^2
sigma_squared = 1 / (idata3.posterior['tau'].values[:, :, np.newaxis] * tau_i)

# hat matrix diagonals
n = sigma_squared.shape[0] * sigma_squared.shape[1]
hat_diag = np.zeros((n, len(y)))
for i in range(n):
    sample_idx = np.unravel_index(i, (sigma_squared.shape[0], sigma_squared.shape[1]))
    D = np.diag(sigma_squared[sample_idx])
    hat_diag[i] = hat_matrix_diag(X, D)

# outlier and non - outlier influence
outlier_hat_diag = hat_diag[:, outlier_i]
non_outlier_hat_diag = hat_diag[:, ~outlier_i]
print(outlier_hat_diag)
print(non_outlier_hat_diag)
print(outlier_indices)
```

```python
#posterior distributions of 1/sigma_i^2
az.plot_trace(1/sigma_squared)
#posterior distributions of tau_i
az.plot_trace(idata3, var_names=['tau_i'])
```
